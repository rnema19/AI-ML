I would start with linear regression model as we want to predict the values from the given data, so we choose regression and not classification.
Here we have a simple relation which can be seen and the values(dispersion) could be fit by a straight line, so we don't choose complex regression models like multiple linear regression, decision tree regression etc.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#sample data taken from kaggle
df = pd.read_csv("score.csv")
df.sample(10)
```

```python
plt.scatter(df["Hours"],df["Scores"])
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,1:2]
```

```python
#X
```

```python
#y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#choosing training data as 40% of the total data
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.4,random_state=42)
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
#checking how accurate our model is
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
lr.predict(X_test.iloc[1].values.reshape(1,1))
```

```python
lr.predict(X_test.iloc[2].values.reshape(1,1))
```

```python
for i in range(5):
    predicted = lr.predict(X_test.iloc[i].values.reshape(1,1))
    actual = y_test.iloc[i]
    print(f"Hours: {X_test.iloc[i,0]}, Predicted Score: {predicted[0,0]}, Actual Score: {actual[0]}")
```

```python
#we check the fit line
plt.scatter(df["Hours"],df["Scores"])
plt.plot(X_test,lr.predict(X_test),color="black",linewidth=3)
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
```
